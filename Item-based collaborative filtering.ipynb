```python
import math
import re
from collections import Counter

WORD = re.compile(r"\w+")


def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in list(vec1.keys())])
    sum2 = sum([vec2[x] ** 2 for x in list(vec2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator


def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)

# def get_similarity(text1,text2):

#   vector1 = text_to_vector(str(text1))
#   vector2 = text_to_vector(str(text2))
#   cosine = get_cosine(vector1, vector2)
#   return cosine

```

```python
from sentence_transformers import SentenceTransformer, util

# Load a pre-trained SBERT model
model = SentenceTransformer('all-mpnet-base-v2')





def get_similarity(sentence1, sentence2):
  # Get the embeddings for each sentence
  embedding1 = model.encode(str(sentence1), convert_to_tensor=True)
  embedding2 = model.encode(str(sentence2), convert_to_tensor=True)

  # Calculate the cosine similarity between the embeddings
  cosine_similarity = util.cos_sim(embedding1, embedding2)
  return float(cosine_similarity.item())
```

```python
item_list = items['item_id'].tolist()   #list of all items
def recommendation(user_id):

  #list of items that this user has seen
  seen_list = ratings[ ratings['user_id']==user_id]['item_id'].tolist()
  #list of items that this user hasn't seen
  unseen_list = [item for item in item_list if item not in seen_list]
  dict = {} #dictionary for unseen items with recommendation score

  for unseen in unseen_list:
    total_similarity, total_score = 0, 0
    for seen in seen_list:
      rating = int(ratings[ (ratings['user_id']== user_id) & (ratings['item_id']== seen) ]['rating'])
      similarity = item_similarity(unseen,seen)
      total_score+= rating*similarity
      total_similarity += similarity
    dict[unseen]= float(total_score / total_similarity )


  return dict

```

```python
from tqdm import tqdm
def calculate_ratings(user_id,item_id):

  #list of items that this user has seen
  seen_list = ratings[ ratings['user_id']==user_id]['item_id'].tolist()
  #list of items that this user hasn't seen
  #unseen_list = [item for item in item_list if item not in seen_list]


  total_similarity, total_score = 0, 0
  for seen in seen_list:
    rating = list(ratings[ (ratings['user_id']== user_id) & (ratings['item_id']== seen) ]['rating'])[0]
    similarity = item_similarity(item_id,seen)
    total_score+= rating*similarity
    total_similarity += similarity
  return float(total_score / total_similarity )


```

#Evaluate

```python
import math
import numpy as np

truth = test['rating'].tolist()
predict = []
for i in tqdm(range(test.shape[0])):
  user , item = test['user_id'][i], test['item_id'][i]
#  print(user,item)
  predict.append(calculate_ratings(user,item))



MSE = np.square(np.subtract(truth,predict)).mean()

RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```
